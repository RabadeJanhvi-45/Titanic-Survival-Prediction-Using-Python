```python
final_df=pd.concat([dummy_cat,titanic['Age']],axis=1)#concat is a vertical join
```

```python
final_df=final_df.drop(['Sex_male'],axis=1)#concat is a vertical join
```

```python
#Prepare x and y for final analysis
x=final_df
y=titanic['Survived']
```

```python
#Split the dataset into training and testing
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=.3,random_state=45)
```

```python
#Build the model
from sklearn.ensemble import RandomForestClassifier
from sklearn.ensemble import AdaBoostClassifier

#instamce of classifier
clf1=AdaBoostClassifier(n_estimators=100,learning_rate=1.0,random_state=45)

#train the model
model=clf1.fit(x_train,y_train)

#test the model
y_pred=model.predict(x_test)
```

```python
#compute the accuracy of the model
from sklearn.metrics import accuracy_score
print("Accuracy : ",(accuracy_score(y_test,y_pred)*100))
```
